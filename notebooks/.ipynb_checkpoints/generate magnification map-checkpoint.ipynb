```python
import TripleLensing
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
import time
TRIL = TripleLensing.TripleLensing()
```

```python
from mpl_toolkits import axes_grid1
def add_colorbar(im, aspect=20, pad_fraction=0.5, **kwargs):
    """Add a vertical color bar to an image plot."""
    divider = axes_grid1.make_axes_locatable(im.axes)
    width = axes_grid1.axes_size.AxesY(im.axes, aspect=1./aspect)
    pad = axes_grid1.axes_size.Fraction(pad_fraction, width)
    current_ax = plt.gca()
    cax = divider.append_axes("right", size=width, pad=pad)
    plt.sca(current_ax)
    return im.axes.figure.colorbar(im, cax=cax, **kwargs)
```

```python
secnum = 45 # divide the source bondary into how many parts
basenum = 2 # the number density of sampled dots among each part
quaderr_Tol = 1.5 # the Quadrupole test tolerance. If this value is larger, the code tend to use point source magnification
relerr_Tol = 1e-3 # the relative error tolerance for magnification
```

```python
#set up lens system
#fractional lens masses: m1, m2, m3
mlens = [0.968738798957637, 0.028093425169771, 0.003167775872591]
#lens positions: x1, y1, x2, y2, x3, y3
zlens = [-0.039343051506317, 0, 1.356656948493683, 0, 0.638936196010800, -0.950873946634155]
rho = 1e-3
```

```python
# generate grid of the magnification map
nmesh = 400
xmin, xmax, ymin, ymax = -2, 2, -2, 2
x0 = np.linspace(xmin,xmax,nmesh+1)
y0 = np.linspace(ymin,ymax,nmesh+1)
x,y = np.meshgrid(x0,y0)
x = x.flatten()
y = y.flatten()
```

```python
# calculate the magnification map
tim0 = time.time()
mags = TRIL.TriLightCurve(mlens, zlens, x, y, rho, secnum, basenum, quaderr_Tol, relerr_Tol)
print("time in generating mags = ", time.time() - tim0)
```

```python
mags = np.array(mags).reshape(nmesh+1,-1)
```

```python
# calculate critical curve and caustics

# x1 x2 x3, y1, y2, y3
resxy = TRIL.outputCriticalTriple_list(mlens, zlens[::2] + zlens[1::2], len(mlens), 2000)
critical = []
caustics = []
numcrit = int(resxy[0])
for i in range(numcrit):
    critical.append( [ resxy[2*i+1], resxy[2*i+2] ] )
offset = 2*numcrit + 1
numcaus = int( resxy[offset] )
for i in range(numcaus):
    caustics.append([resxy[offset+2*i+1],resxy[offset+2*i+2]])
causx = np.array([xy[0]for xy in caustics])
causy = np.array([xy[1]for xy in caustics])
criticalx = [xy[0]for xy in critical]
criticaly = [xy[1]for xy in critical]
```

```python
#plot 

f = plt.figure(figsize=(8,8),dpi=100)
gs = gridspec.GridSpec(1,1)
ax = plt.subplot(gs[0])
plt.subplots_adjust(top = 0.9, bottom = 0.08, right = 0.95, left = 0.1, hspace = 0.1, wspace = 0.2)

im = ax.imshow((np.log10(mags)), extent = [xmin, xmax, ymin, ymax],origin = "lower", cmap = "jet")
cb = add_colorbar(im)
cb.set_label(r"log $\mu$", fontsize = 14)

ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

ax.scatter(zlens[0], zlens[1], marker = "+",s = 200, color = "r")
ax.scatter(zlens[2], zlens[3], marker = "+",s = 200, color = "r")
ax.scatter(zlens[4], zlens[5], marker = "+",s = 200, color = "r")
ax.scatter(causx,causy, marker='.', s= 1,color='r')
ax.scatter(criticalx,criticaly, marker='.', s= 1,color='k')

ax.set_xlabel(r"$x/\theta_E$", fontsize = 14);
ax.set_ylabel(r"$y/\theta_E$", fontsize = 14);
```
